**Year and region heatmap**

the attack recivers region conatins multi valued lis like strings and some semicolon separated, so to analyse region trents it is normalised and exploded.

```python
def parse_regions(x):
    if pd.isna(x):
        return []

    # Split on semicolon if multiple lists exist
    parts = str(x).split(";")
    regions = []

    for p in parts:
        p = p.strip()
        # Remove brackets
        p = p.strip("[]")
        # Split on commas
        items = p.split(",")
        for item in items:
            clean = item.replace("'", "").replace('"', "").strip()
            if clean and clean.lower() != "not available":
                regions.append(clean)

    return list(set(regions))  # removed duplicates
```

```python
df_regions = df.copy()

df_regions["receiver_regions"] = df_regions["receiver_regions"].apply(parse_regions)

df_regions = df_regions.explode("receiver_regions")

df_regions = df_regions.dropna(subset=["receiver_regions", "year"])
```

```python
heat = (
    df_regions
    .groupby(["year", "receiver_regions"])
    .size()
    .reset_index(name="count")
)
```

```python
px.density_heatmap(
    heat,
    x="year",
    y="receiver_regions",
    z="count",
    title="Cyber Incidents by Year and Region",
    labels={"year": "Year", "receiver_regions": "Region", "count": "Number of Incidents"}
)
```

The receiver_regions variable contained complex, multi-valued entries, including multiple region lists stored as semicolon-separated strings.
A custom parsing function was implemented to normalise this field by extracting individual regions and exploding the data so that each incident–region relationship could be analysed separately.
This approach improves interpretability while acknowledging that a single cyber incident may affect multiple regions simultaneously.
